# №1.50


Пусть X и Y - это случайные моменты прихода двух лиц.\
|Х - У| <= 0.25\
Условие встречи можно записать в виде неравенства |х-у| >= 0.25 и |x-y| <= 0.25\
то есть на графике можно построить два уравнения: Y = x + 0.25 и y = x - 0.25\
Вероятность встречи - это площадь фигуры между двумя прямыми\ 
Область вне прямых состоит из двух треугольников с основанием и высотой = 0.75\
отсюда площади этих двух треугольников равны 0.75*0.75 = 0.5625\
А значит вероятность того, что лица встретятся равна 1 - 0.5625 = 0.4375



# №1.51
\
Найдем вероятность попадания как отношение площади круга к площади квадрата\
Площадь круга $$\pi\cdot r^{2}$$\
 Площадь квадрата $$2\cdot r^{2}$$\
 Тогда вероятность попадания будет вычисляться как:\
 $$\frac{2\cdot r^{2}}{\pi r^{2}}=\frac{2}{\pi}$$


```python
import math as m
a = 2 / m.pi
print(round(a, 4))
```
